```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import kagglehub
from collections import Counter
```

```python
DATASET_PATH = kagglehub.dataset_download(
    "tuanai/traffic-signs-dataset"
)
print("Dataset:", DATASET_PATH)
```

```python
IMG_EXTS = (".jpg", ".jpeg", ".png")

def list_images(base_dir):
    paths = []
    for root, _, files in os.walk(base_dir):
        for f in files:
            if f.lower().endswith(IMG_EXTS):
                paths.append(os.path.join(root, f))
    return paths

rows = list_images(DATASET_PATH)
print("Liczba obrazów:", len(rows))
```

```python
def red_mask(hsv):
    m1 = cv2.inRange(hsv, (0, 50, 40), (10, 255, 255))
    m2 = cv2.inRange(hsv, (170, 50, 40), (180, 255, 255))
    return cv2.bitwise_or(m1, m2)

def yellow_mask(hsv):
    m1 = cv2.inRange(hsv, (10, 50, 50), (35, 255, 255))
    m2 = cv2.inRange(hsv, (10, 30, 30), (40, 255, 255))
    return cv2.bitwise_or(m1, m2)

def blue_mask(hsv):
    lower = np.array([95, 80, 80])
    upper = np.array([135, 255, 255])
    return cv2.inRange(hsv, lower, upper)

```

```python
def clean_mask(mask):
    k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k, 2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, k, 1)
    return mask
```

```python
def find_sign_contour(mask):
    cnts, _ = cv2.findContours(
        mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )
    if not cnts:
        return None, 0

    best = max(cnts, key=cv2.contourArea)
    area = cv2.contourArea(best)
    return best, area
```

```python
def analyze_contour(c):
    peri = cv2.arcLength(c, True)
    if peri == 0:
        return None

    area = cv2.contourArea(c)
    approx = cv2.approxPolyDP(c, 0.03 * peri, True)
    verts = len(approx)
    circ = 4 * np.pi * area / (peri * peri)

    return verts, circ, approx
```

```python
def polygon_angle_std(approx):
    pts = approx.reshape(-1, 2)
    n = len(pts)
    if n < 3:
        return 999

    angles = []
    for i in range(n):
        p0 = pts[i - 1]
        p1 = pts[i]
        p2 = pts[(i + 1) % n]

        v1 = p0 - p1
        v2 = p2 - p1

        cosang = np.dot(v1, v2) / (
            np.linalg.norm(v1) * np.linalg.norm(v2) + 1e-6
        )
        cosang = np.clip(cosang, -1, 1)
        angle = np.degrees(np.arccos(cosang))
        angles.append(angle)

    return np.std(angles)
```

```python
def overlay_mask(img, mask, color=(0,255,0), alpha=0.4):
    overlay = img.copy()
    color_layer = np.zeros_like(img)
    color_layer[mask > 0] = color
    cv2.addWeighted(color_layer, alpha, overlay, 1 - alpha, 0, overlay)
    return overlay
```

```python
def classify_sign(img):
    img = cv2.resize(img, (256,256))
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # ---------- ZAKAZ (czerwone koło) ----------
    mask_r = clean_mask(red_mask(hsv))
    c, area = find_sign_contour(mask_r)

    if c is not None and area > 1500:
        res = analyze_contour(c)
        if res is not None:
            _, circ, approx = res
            if circ >= 0.85:
                return "ZAKAZ", approx, mask_r

    # ---------- NAKAZ (niebieskie koło) ----------
    mask_b = clean_mask(blue_mask(hsv))
    c, area = find_sign_contour(mask_b)

    if c is not None and area > 1500:
        res = analyze_contour(c)
        if res is not None:
            _, circ, approx = res
            if circ >= 0.85:
                return "NAKAZ", approx, mask_b

    # ---------- TRIANGLE (żółty) ----------
    mask_y = clean_mask(yellow_mask(hsv))
    c, area = find_sign_contour(mask_y)

    if c is not None and area > 1500:
        res = analyze_contour(c)
        if res is not None:
            v, circ, approx = res
            if v <= 4 and circ < 0.75:
                return "OSTRZEGAWCZY", approx, mask_y

    return "UNKNOWN", None, None
```

```python
def show_examples(rows, N=5):
    sample = np.random.choice(rows, min(N, len(rows)), replace=False)
    fig, axs = plt.subplots(3, len(sample), figsize=(4*len(sample), 9))

    for i, path in enumerate(sample):
        img = cv2.imread(path)
        if img is None:
            continue

        img = cv2.resize(img, (256,256))
        label, _, mask = classify_sign(img)

        axs[0,i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axs[0,i].set_title("Original")
        axs[0,i].axis("off")

        axs[1,i].imshow(mask if mask is not None else np.zeros((256,256)), cmap="gray")
        axs[1,i].set_title("Mask")
        axs[1,i].axis("off")

        over = overlay_mask(img, mask) if mask is not None else img
        axs[2,i].imshow(cv2.cvtColor(over, cv2.COLOR_BGR2RGB))
        axs[2,i].set_title(label)
        axs[2,i].axis("off")

    plt.tight_layout()
    plt.show()

```

```python
def plot_shape_complexity(rows, N=200):
    verts = []
    sample = np.random.choice(rows, min(N, len(rows)), replace=False)

    for path in sample:
        img = cv2.imread(path)
        if img is None:
            continue

        img = cv2.resize(img, (256,256))
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

        for mask in [
            clean_mask(red_mask(hsv)),
            clean_mask(yellow_mask(hsv)),
            clean_mask(blue_mask(hsv))
        ]:
            c, _ = find_sign_contour(mask)
            if c is None:
                continue

            res = analyze_contour(c)
            if res:
                v, _, _ = res
                verts.append(v)

    plt.figure(figsize=(7,4))
    plt.hist(verts, bins=range(3,16), edgecolor="black")
    plt.xlabel("Liczba wierzchołków")
    plt.ylabel("Liczność")
    plt.title("Rozkład złożoności kształtów")
    plt.tight_layout()
    plt.show()

```

```python
def plot_prediction_distribution(rows, N=200):
    preds = []
    sample = np.random.choice(rows, min(N, len(rows)), replace=False)

    for path in sample:
        img = cv2.imread(path)
        if img is None:
            continue

        label, _, _ = classify_sign(img)
        preds.append(label)

    c = Counter(preds)
    plt.figure(figsize=(7,4))
    plt.bar(c.keys(), c.values())
    plt.ylabel("Liczność")
    plt.title("Rozkład predykcji klas")
    plt.tight_layout()
    plt.show()

```

```python
show_examples(rows, N=5)
plot_shape_complexity(rows)
plot_prediction_distribution(rows)

```
